# Open dataframe

```python
import numpy as np
import scipy as sp
import pandas as pd

import rpy2.rinterface
%load_ext rpy2.ipython
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
%%R -o df

df <- get(load("dpb.Rdata"))
```

```python
df = df.rename(columns={'TARGET':'y'})
```

```python
df = df.drop(columns=['AMT_REQ_CREDIT_BUREAU_TOTAL'])
```

```python
np.array([len(df[col].unique()) for col in df.columns])
```

```python
df_boolean = df.iloc[:, np.where(np.array([len(df[col].unique()) for col in df.columns]) == 2)[0]].copy()
```

```python
for col in df_boolean.columns:
    df_boolean[col] = df_boolean[col] - 1
df_boolean
```

```python
categorcial_names = df.iloc[:, np.where(np.array([len(df[col].unique()) for col in df.columns]) > 2)[0]].iloc[:, [0,6,7,8,9,10,15,16,18,19,21,23,24,25,26]].columns
```

```python
df_categorical = df[categorcial_names].copy()
for col in df_categorical.columns:
    df_categorical[col] = df_categorical[col].astype("category")
df_categorical
```

```python
lista_colonne_restanti = []
for col in df.columns:
    if col in df_boolean.columns:
        continue
    if col in df_categorical.columns:
        continue
    if len(df[col].unique()) < 2:
        continue
    lista_colonne_restanti.append(col)
```

```python
lista_colonne_restanti
```

```python
df_numerical = df[lista_colonne_restanti].copy()
```

```python
for col in df_numerical.columns:
    df_numerical[col] = (df_numerical[col]-df_numerical[col].mean())/df_numerical[col].std()
```

```python
df_testset_raw = pd.read_csv("testset.csv")
```

```python
df_testset_raw = df_testset_raw.rename(columns={'TARGET':'y'})
```

```python
df_test = df_testset_raw.loc[:,df_boolean.columns.union(df_numerical.columns)].copy()
df_test_cat = df_testset_raw.loc[:,categorcial_names].copy()
for col in df_test_cat.columns:
    df_test_cat[col] = df_test_cat[col].astype("category")
df_test = df_test.join(pd.get_dummies(df_test_cat, drop_first=False))
```

```python
df_test
```

# Analysis

```python
df_boolean_uno = df_boolean[df_boolean['y'] == 1]
indici_uno_training = df_boolean_uno.sample(n = 800).index
df_boolean_uno = df_boolean_uno.drop(indici_uno_training)
indici_uno_calibration = df_boolean_uno.index
```

```python
df_boolean_zero = df_boolean[df_boolean['y'] == 0]
indici_zero_training = df_boolean_zero.sample(n = 13774).index
df_boolean_zero = df_boolean_zero.drop(indici_zero_training)
indici_zero_calibration = df_boolean_zero.index
```

```python
indici_training = indici_zero_training.union(indici_uno_training)
indici_calibration = indici_zero_calibration.union(indici_uno_calibration)
```

```python
df_calibration = df_boolean.loc[indici_calibration,:].join(pd.get_dummies(df_categorical.loc[indici_calibration,:], drop_first=False))
df_calibration = df_calibration.join(df_numerical.loc[indici_calibration,:])
df_calibration
```

```python
df_test = df_test.loc[:,df_calibration.columns]
```

df_test = df_boolean.loc[indici_test,:].join(pd.get_dummies(df_categorical.loc[indici_test,:], drop_first=False))
df_test = df_test.join(df_numerical.loc[indici_test,:])
df_test

```python
df_test_wrong = df_test.copy()

df_test_wrong['y'] = (1 - df_test['y']).copy()
```

```python
from sklearn.neural_network import MLPClassifier

clf = MLPClassifier(solver='lbfgs', alpha=1e-5,
                    hidden_layer_sizes=[165,165,165,165], random_state=1)
```

```python
clf.fit(df_training.iloc[:,1:], df_training.iloc[:,0])
```

```python
sum(clf.predict(df_calibration.iloc[:,1:]))
```

```python
N_training = df_training.shape[0]
N_calibration = df_calibration.shape[0]
N_test = df_test.shape[0]
```

```python
probs = np.transpose(clf.predict_proba(df_calibration.iloc[:,1:]))
df_calibration['NCM'] = [probs[1-df_calibration.iloc[i,0],i] for i in range(N_calibration)]
df_calibration = df_calibration.sort_values('NCM')
df_calibration
```

```python
probs = np.transpose(clf.predict_proba(df_test.iloc[:,1:]))
df_test['NCM'] = [probs[1-df_test.iloc[i,0],i] for i in range(N_test)]
df_test['p_val'] = 1 - np.array( [ np.searchsorted(df_calibration['NCM'].to_numpy(), val)+1 for val in df_test['NCM'] ] )/N_calibration
```

```python
df_test_wrong['NCM'] = [probs[1-df_test_wrong.iloc[i,0],i] for i in range(N_test)]
df_test_wrong['p_val'] = 1 - np.array( [ np.searchsorted(df_calibration['NCM'].to_numpy(), val)+1 for val in df_test_wrong['NCM'] ] )/N_calibration
```

```python
(df_test['p_val'] < 0.05).mean()
```

```python
def true_positive(p):
    return (1 - (df_test[df_test['y'] == 1]['p_val'] < p).mean())
def false_positive(p):
    return (1 - (df_test_wrong[df_test['y'] == 0]['p_val'] < p).mean())
```

```python
pd.DataFrame([(true_positive(p),false_positive(p)) for p in np.linspace(0,0.1,1001)], columns=['true_positive','false_positive']).plot(kind='line', x='false_positive', y='true_positive')
plt.plot(np.linspace(0,1,101),np.linspace(0,1,101))
```
