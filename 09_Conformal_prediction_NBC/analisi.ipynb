```python
df_boolean = df.iloc[:, np.where(np.array([len(df[col].unique()) for col in df.columns]) == 2)[0]].copy()
```

```python
for col in df_boolean.columns:
    df_boolean[col] = df_boolean[col] - 1
df_boolean
```

```python
categorcial_names = df.iloc[:, np.where(np.array([len(df[col].unique()) for col in df.columns]) > 2)[0]].iloc[:, [0,6,7,8,9,10,15,16,18,19,21,23,24,25,26]].columns
```

```python
df_categorical = df[categorcial_names].copy()
for col in df_categorical.columns:
    df_categorical[col] = df_categorical[col].astype("category")
df_categorical
```

# Analysis

## ATTENZIONE! È IMPORTANTE CHE Y SIA LA PRIMA ENTRATA NELLA TABELLA!!!!

```python
df_boolean_uno = df_boolean[df_boolean['y'] == 1]
indici_uno_training = df_boolean_uno.sample(n = 550).index
df_boolean_uno = df_boolean_uno.drop(indici_uno_training)
indici_uno_calibration = df_boolean_uno.sample(n = 275).index
df_boolean_uno = df_boolean_uno.drop(indici_uno_calibration)
indici_uno_test = df_boolean_uno.index
```

```python
df_boolean_zero = df_boolean[df_boolean['y'] == 0]
indici_zero_training = df_boolean_zero.sample(n = 9182).index
df_boolean_zero = df_boolean_zero.drop(indici_zero_training)
indici_zero_calibration = df_boolean_zero.sample(n = 4592).index
df_boolean_zero = df_boolean_zero.drop(indici_zero_calibration)
indici_zero_test = df_boolean_zero.index
```

```python
indici_training = indici_zero_training.union(indici_uno_training)
indici_calibration = indici_zero_calibration.union(indici_uno_calibration)
indici_test = indici_zero_test.union(indici_uno_test)
```

```python
N_training = 9732
N_calibration = 4867
N_test = 4868

m = 30
m_cat = np.array([len(df[col].unique()) for col in df_categorical.columns])

df_training = df_boolean.loc[indici_training,:].copy()
df_cat_training = df_categorical.loc[indici_training,:].copy()
df_calibration = df_boolean.loc[indici_calibration,:].copy()
df_cat_calibration = df_categorical.loc[indici_calibration,:].copy()
df_test = df_boolean.loc[indici_test,:].copy()
df_cat_test = df_categorical.loc[indici_test,:].copy()
```

```python
df_test_wrong = df_test.copy()

df_test_wrong['y'] = (1 - df_test['y']).copy()
```

```python
medie = df_training.mean()
```

```python
df_cat_training_dummy = pd.get_dummies(df_cat_training, drop_first=False)
df_cat_training_dummy['y'] = df_training['y']
df_cat_calibration_dummy = pd.get_dummies(df_cat_calibration, drop_first=False)
df_cat_calibration_dummy['y'] = df_calibration['y']
df_cat_test_dummy = pd.get_dummies(df_cat_test, drop_first=False)
df_cat_test_dummy['y'] = df_test['y']
```

```python
freqs_cond_on_y = df_training.groupby('y').mean()
freqs_cond_on_y[freqs_cond_on_y == 0] = 1/N_training
freqs_cond_on_y
```

```python
freqs_cat_cond_on_y = df_cat_training_dummy.groupby('y').mean()
freqs_cat_cond_on_y[freqs_cat_cond_on_y == 0] = 1/N_training
freqs_cat_cond_on_y
```

```python
log_freqs_cond_on_y = np.log(freqs_cond_on_y.to_numpy())
log_freqs_cond_on_y
```

```python
log_freqs_cat_cond_on_y = np.log(freqs_cat_cond_on_y.to_numpy())
log_freqs_cat_cond_on_y
```

```python
log_prob_test = np.matmul(log_freqs_cond_on_y, np.transpose(df_calibration.to_numpy()[:,0:m])) + \
				np.array([np.log(1 - medie['y'])*np.ones(N_calibration), np.log(medie['y'])*np.ones(N_calibration)]) + \
				np.matmul(log_freqs_cat_cond_on_y, np.transpose(df_cat_calibration_dummy.to_numpy()[:,0:-1]))

normalizzazione = sp.special.logsumexp(log_prob_test, axis=0)
log_prob_test = log_prob_test - normalizzazione
probs = np.exp(log_prob_test)
df_calibration['NCM'] = [probs[1-df_calibration.iloc[i,0],i] for i in range(N_calibration)]
df_calibration = df_calibration.sort_values('NCM')
df_calibration
```

```python
import matplotlib.pyplot as plt
plt.plot(df_calibration['NCM'], np.linspace(0,1,N_calibration,endpoint=False))
```

```python
def last(arr):
    if arr.size == 0:
        return 0
    return arr[-1]
```

```python
log_prob_test = np.matmul(log_freqs_cond_on_y, np.transpose(df_test.to_numpy()[:,0:m])) + \
				np.array([np.log(1 - medie['y'])*np.ones(N_test), np.log(medie['y'])*np.ones(N_test)]) + \
				np.matmul(log_freqs_cat_cond_on_y, np.transpose(df_cat_test_dummy.to_numpy()[:,0:-1]))


normalizzazione = sp.special.logsumexp(log_prob_test, axis=0)
log_prob_test = log_prob_test - normalizzazione
probs = np.exp(log_prob_test)
df_test['NCM'] = [probs[1-df_test.iloc[i,0],i] for i in range(N_test)]
df_test['p_val'] = 1 - np.array( [ np.searchsorted(df_calibration['NCM'].to_numpy(), val)+1 for val in df_test['NCM'] ] )/N_calibration
```

```python
(df_test['p_val'] < 0.05).mean()
```

```python
log_prob_test = np.matmul(log_freqs_cond_on_y, np.transpose(df_test_wrong.to_numpy()[:,0:m])) + \
				np.array([np.log(1 - medie['y'])*np.ones(N_test), np.log(medie['y'])*np.ones(N_test)]) + \
				np.matmul(log_freqs_cat_cond_on_y, np.transpose(df_cat_test_dummy.to_numpy()[:,0:-1]))


normalizzazione = sp.special.logsumexp(log_prob_test, axis=0)
log_prob_test = log_prob_test - normalizzazione
probs = np.exp(log_prob_test)
df_test_wrong['NCM'] = [probs[1-df_test_wrong.iloc[i,0],i] for i in range(N_test)]
df_test_wrong['p_val'] = 1 - np.array( [ np.searchsorted(df_calibration['NCM'].to_numpy(), val)+1 for val in df_test_wrong['NCM'] ] )/N_calibration
```
